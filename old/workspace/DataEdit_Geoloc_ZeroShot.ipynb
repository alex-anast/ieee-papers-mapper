```python
# for LOCATION (API)
def loc(df):
    # Create lists with info and index for pipeline
    entity_info = []
    entity_index = []
    for row_index in range(len(df)):
        if isinstance(df["authors.authors"][row_index], float):
            continue

        list_affiliations = df["authors.authors"][row_index].split(";")
        list_affiliations = list(dict.fromkeys(list_affiliations))  # Remove duplicates

        for entity in list_affiliations:
            if (
                "https" in entity or "full_name" in entity
            ):  # it gets included as an "Error" below
                continue
            entity = "".join([ch for ch in entity if ch.isalpha() or ch in " _-"])
            entity_info.append(entity)
            entity_index.append(row_index)

    print("# Countries to find : %d" % len(entity_info))
    start = time.time()

    # PIPELINE
    country = []
    country_index = []
    count_uncategorized = 0
    row_index = 0
    for judgement in classifierGPU(
        entity_info,
        candidate_labels,
        hypothesis_template=hypothesis_template,
        batch_size=32,
        multi_label=False,
    ):  # 16 is the last perfect, 32 is sliiightly better
        if judgement["scores"][0] > 0.1:
            country.append(str(judgement["labels"][0]) + "; ")
            # print("Country= %s, Score= %0.3f" % (judgement["labels"][0], judgement["scores"][0]))
        else:
            print("//-----")
            print("FAILED : ", entity_info[row_index], entity_index[row_index])
            country.append("Error; ")
            count_uncategorized += 1
        country_index.append(entity_index[row_index])
        row_index += 1

    # Standardize UK and USA abbs.
    for i in range(len(country)):
        data = country[i]
        data = data[: len(data) - 2]
        if data in ["U K", "UK", "England", "United Kingdom"]:
            country[i] = "UK; "
        elif data in ["USA", "US", "U S", "United States"]:
            country[i] = "USA; "

    # Create Dictionary to combine data per Row ID (FIX DF)
    create_dict = {}
    for i in range(len(country)):
        if country_index[i] not in create_dict:
            create_dict.update({country_index[i]: country[i]})
        else:
            create_dict.update(
                {country_index[i]: create_dict[country_index[i]] + country[i]}
            )

    # REMOVE DUPLICATE COUNTRIES
    remove_duplicates = list(create_dict.items())
    remove_duplicates_id = list(create_dict.keys())
    for entity_id in range(len(remove_duplicates)):
        remove_duplicates[entity_id] = remove_duplicates[entity_id][1].split("; ")
        remove_duplicates[entity_id] = "; ".join(
            list(
                dict.fromkeys(
                    remove_duplicates[entity_id][
                        : len(remove_duplicates[entity_id]) - 1
                    ]
                )
            )
        )
        if len(remove_duplicates[entity_id]) == 0:
            remove_duplicates[entity_id] = "Error"
    # THE ONES THAT WERE SKIPPED
    for i in range(len(df)):
        if i not in remove_duplicates_id:
            df.loc[i, "Country"] = "Error"
    # FILL COUNTRIES
    for i in range(len(remove_duplicates_id)):
        df.loc[remove_duplicates_id[i], "Country"] = remove_duplicates[i]

    # Fix Error written in Uncategorized ones:
    for i in range(len(df)):
        if "Error" in df.loc[i, "Country"]:
            if not df.loc[i, "Country"] == "Error":
                df.loc[i, "Country"] = df.loc[i, "Country"].replace("; Error", "")
    # ..and rename to Uncategorized
    for i in range(len(df)):
        df.loc[i, "Country"] = df.loc[i, "Country"].replace("Error", "Uncategorized")

    # Set Continent for every country
    df["Continent"] = NaN
    for i in range(len(df)):
        if df.loc[i, "Country"] == "Uncategorized":
            df.loc[i, "Continent"] = "Uncategorized"
            continue
        extract = df.loc[i, "Country"].split("; ")
        data = ""
        for j in extract:
            for k in range(len(countries)):
                if j in countries[k]:
                    data += list_continents[k] + "; "
        df.loc[i, "Continent"] = data[: len(data) - 2]

    m, s = divmod(time.time() - start, 60)
    print("#Papers = %d, Time Total Elapsed: %d min %0.2f sec" % (i + 1, m, s))
    print("Time Elapsed : %0.2f" % (time.time() - start))
    print("# Uncategorized : %d" % count_uncategorized)

    return df


df = loc(df)
```
